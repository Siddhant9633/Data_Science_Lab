```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Iris.csv")

# Display basic info
print("Columns:", df.columns)
print("\nFirst 5 rows:\n", df.head())

# -----------------------------------
# 1. Filtering Data
# -----------------------------------

# Petal Length > 4
filtered = df[df['PetalLengthCm'] > 4]
print("\nFiltered (PetalLength > 4):\n", filtered.head())

# Only Setosa
setosa = df[df['Species'] == 'Iris-setosa']
print("\nSetosa Sample:\n", setosa.head())

# -----------------------------------
# 2. Grouping + Aggregation
# -----------------------------------

grouped = df.groupby('Species')

print("\nMean values by Species:\n")
print(grouped.mean(numeric_only=True))

# -----------------------------------
# 3. Visualization (IMPORTANT FOR MARKS)
# -----------------------------------

# Scatter Plot
plt.figure()
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.scatter(subset['PetalLengthCm'], subset['PetalWidthCm'], label=species)

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Width")
plt.legend()
plt.show()

# Histogram
plt.figure()
df['SepalLengthCm'].hist()
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()

# -----------------------------------
# 4. Insights
# -----------------------------------

print("\nInsights:")
print("1. Setosa has smallest petals.")
print("2. Virginica has largest petals.")
print("3. Versicolor lies in between.")
```
