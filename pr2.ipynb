```python
# Graph represented as adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# BFS Implementation
def bfs(graph, start):
    visited = []
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])
    return visited

# DFS Implementation
def dfs(graph, start, visited=None):
    if visited is None:
        visited = []

    if start not in visited:
        visited.append(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)
    return visited

# Execution
print("BFS Traversal:", bfs(graph, 'A'))
print("DFS Traversal:", dfs(graph, 'A'))
```

```python
# A* Algorithm Example

import heapq

# Graph with costs
graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 1), ('E', 5)],
    'C': [('F', 2)],
    'D': [],
    'E': [('F', 1)],
    'F': []
}

# Heuristic values (estimated cost to goal)
heuristic = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 1,
    'E': 1,
    'F': 0
}

def a_star(start, goal):
    pq = [(0 + heuristic[start], 0, start, [])]  # (f, g, node, path)
    visited = set()

    while pq:
        f, g, node, path = heapq.heappop(pq)

        if node in visited:
            continue

        path = path + [node]
        visited.add(node)

        if node == goal:
            return path, g

        for neighbor, cost in graph[node]:
            if neighbor not in visited:
                heapq.heappush(pq, (g + cost + heuristic[neighbor],
                                    g + cost,
                                    neighbor,
                                    path))
    return None

# Execution
path, cost = a_star('A', 'F')
print("A* Path:", path)
print("Total Cost:", cost)
```

```python
# Minimax Algorithm Implementation

def minimax(depth, node_index, is_max, values, height):
    # If leaf node reached
    if depth == height:
        return values[node_index]

    if is_max:
        return max(
            minimax(depth + 1, node_index * 2, False, values, height),
            minimax(depth + 1, node_index * 2 + 1, False, values, height)
        )
    else:
        return min(
            minimax(depth + 1, node_index * 2, True, values, height),
            minimax(depth + 1, node_index * 2 + 1, True, values, height)
        )

# Example leaf node values
values = [3, 5, 6, 9, 1, 2, 0, -1]

import math
height = int(math.log2(len(values)))

# Execution
result = minimax(0, 0, True, values, height)
print("Optimal Value (Minimax):", result)
```
