```python
# Program to compute factorial of a number

num = int(input("Enter a number: "))
fact = 1

if num < 0:
    print("Factorial not defined for negative numbers")
else:
    for i in range(1, num + 1):
        fact *= i
    print("Factorial of", num, "is", fact)
```

```python
# Program to display multiplication table

num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# Program to check odd or even

num = int(input("Enter a number: "))

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

```python
# Program to check palindrome number

num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("The number is a Palindrome")
else:
    print("The number is NOT a Palindrome")
```

```python
# Simple Expert System for career suggestion

print("Answer the following questions with yes/no")

likes_math = input("Do you like Mathematics? ").lower()
likes_biology = input("Do you like Biology? ").lower()
likes_computers = input("Do you like Computers? ").lower()

if likes_computers == "yes" and likes_math == "yes":
    print("Suggested Career: Software Engineer / Data Scientist")
elif likes_biology == "yes":
    print("Suggested Career: Doctor / Biotechnologist")
elif likes_math == "yes":
    print("Suggested Career: Engineer / Statistician")
else:
    print("Suggested Career: Arts / Commerce fields")
```
