### Task 1: Generate a normal distribution and plot its histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a normal distribution (e.g., 1000 samples with mean 0, std dev 1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

print("First 10 samples from the normal distribution:", normal_data[:10])
```

```python
plt.figure(figsize=(8, 6))
plt.hist(normal_data, bins=30, density=True, alpha=0.7, color='purple', edgecolor='black')
plt.title('Histogram of a Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Task 2: Compute mean, median, variance, and standard deviation for a dataset

```python
import numpy as np

# Let's use the 'normal_data' generated in the previous step, or define a new one
dataset_for_stats = np.array([12, 15, 18, 20, 22, 25, 28, 30, 35, 40, 42, 45, 50])

print("Dataset:", dataset_for_stats)

# Compute statistics
mean_val = np.mean(dataset_for_stats)
median_val = np.median(dataset_for_stats)
variance_val = np.var(dataset_for_stats)
std_dev_val = np.std(dataset_for_stats)

print(f"\nMean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_dev_val:.2f}")
```
